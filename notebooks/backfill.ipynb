```python
import pandas as pd
import pymysql as pymysql
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cyclop.csv")
```

```python
vals = ['seeing', 'flux', 'r0']
tables = ["seeing_monitor_" + v for v in vals]
tables
```

```python
sql_inputs = {}

for v in vals:
    sql_inputs[v] = list(zip(list(df['epoch_ms']), list(df[v])))

sql_inputs
```

```python
host = "mariadb.mmto.arizona.edu"
database = "measurements"
usr = "mmtstaff"
pw = "multiple"

conn = pymysql.connect(host=host, port=3306, user=usr, passwd=pw, db=database)
```

```python
for v in vals:
    query = f"INSERT INTO seeing_monitor_{v} (timestamp,value) VALUES (%s,%s)"
    cursor = conn.cursor()
    cursor.executemany(query, sql_inputs[v])
    conn.commit()
cursor.close()
conn.close()
```
